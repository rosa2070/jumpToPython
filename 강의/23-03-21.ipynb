````python
# total(1,'2'3,4,5,'6')은 13이 출력되도록 total()를 만드시오.
def total(*args):
    result = 0
    
    for arg in args:
        try:
            result += arg
        except:
            print(f"{arg}는 int가 아닙니다.")
    return result

total(1,'2',3,4,5,'6')

````
